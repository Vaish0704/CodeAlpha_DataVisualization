```python
import matplotlib.pyplot as plt #Author Frequency Bar Chart
import pandas as pd

df = pd.read_csv('/content/quotes_data.csv')

top_authors = df['Author'].value_counts().head(5)

plt.figure()
top_authors.plot(kind='bar')
plt.title("Top 5 Authors by Number of Quotes")
plt.xlabel("Author")
plt.ylabel("Count")
plt.show()
```

```python
df['Quote_Length'] = df['Quote'].apply(len) #Quotes Length Histogram

plt.figure()
plt.hist(df['Quote_Length'], bins=10)
plt.title("Distribution of Quote Lengths")
plt.xlabel("Length")
plt.ylabel("Frequency")
plt.show()
```
